#### .children property

```python
head_tag.children
```

```python
for child in head_tag.children:
    print(child)
```

```python
html_tag.children
```

```python
for child in html_tag.children:
    print(child.name)
```

```python
print("immediate parent", "child")

for child in soup.body.children:
    print(child.parent.name, "\t\t", child.name)
```

```python
for child in head_tag.descendants:
    print(child)
```

```python
for child in html_tag.descendants:
    if child.name is not None:
          print(child.name) 
```

#### .descendants will consider all the lower nodes.

```python
for child in html_tag.descendants:
    if child.name is not None:
          print(child.parent.name, "\t\t", child.name)
```

#### If we want all of the strings in the html tag, we may use html.strings property

```python
html_tag.strings
```

#### It generates a generaor object which we may iterate over

```python
for string in html_tag.strings:
    print(string)
```

#### For avoiding un necessary spaces, we may use stripped_spaces

```python
html_tag.stripped_strings
```

```python
for string in html_tag.stripped_strings:
    print(string)
```
