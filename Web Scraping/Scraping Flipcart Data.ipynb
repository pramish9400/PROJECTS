# Scraping Flipcart Data

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
url = 'https://www.flipkart.com/search?sid=tyy%2C4io&otracker=CLP_Filters&p%5B%5D=facets.brand%255B%255D%3Dvivo&page=1'
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
soup
```

```python
Phone_name = [element.text.strip() for element in soup.find_all('div', {'class':"_4rR01T"})]
```

```python
Phone_name
```

```python
Phone_price = [element.text.strip() for element in soup.find_all("div", class_ = "_30jeq3 _1_WHN1")]
Phone_price
```

### For scraping multiple pages

```python
for i in range(1, 11):
    url = 'https://www.flipkart.com/search?sid=tyy%2C4io&otracker=CLP_Filters&p%5B%5D=facets.brand%255B%255D%3Dvivo&page='+str(i)
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html')
    Phone_Name = [element.text.strip() for element in soup.find_all('div', class_ = "_4rR01T")]
    Phone_price = [element.text.strip() for element in soup.find_all('div', class_ = "_30jeq3 _1_WHN1")]
```

```python
data_scrape = pd.DataFrame({'Model':Phone_Name,
                           'Price':Phone_price})
data_scrape.head()
```
