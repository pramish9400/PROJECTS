## <center> Extracting the List of Links

```python
import webbrowser
import requests
import re 
from bs4 import BeautifulSoup
```

```python
url = "http://www.pluralsight.com"
resp = requests.get(url)
```

```python
soup = BeautifulSoup(resp.text, 'lxml')
soup
```

```python
soup.find_all("a", attrs= {'href':re.compile("^https")})
```

```python
for link in soup.find_all("a", attrs= {'href':re.compile("^https")}):
    print(link['href'])
```

### Extracting non secure links

```python
for link in soup.findAll('a', attrs= {"href":re.compile("^http://")}):
    print(link['href'])
```

```python
for link in soup.findAll('a', href = True):
    print(link['href'])
```

```python
for link in soup.findAll(['a', 'link'], attrs= {"href":re.compile("^http")}):
    print(link['href'])
```

```python
for img in soup.findAll('img'):
    print(img.get('src'))
```
